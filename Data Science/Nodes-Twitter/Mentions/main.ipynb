```python
import json


credentials = {}
credentials['CONSUMER_KEY'] = '#'
credentials['CONSUMER_SECRET'] = '#'
credentials['ACCESS_TOKEN'] = '#'
credentials['ACCESS_SECRET'] = '#'

with open("twitter_credentials.json", "w") as file:
    json.dump(credentials, file)
```

```python
from twython import Twython

# Load credentials from json file
with open("twitter_credentials.json", "r") as file:
    creds = json.load(file)

# Instantiate an object
python_tweets = Twython(creds['CONSUMER_KEY'], creds['CONSUMER_SECRET'])
```

```python
# Create our query
query = {'q': 'to%3AZindiAfrica',
        'result_type': 'recent',
        'count': 100,
        'lang': 'en',
        'include_entities': 'true'         
        }
```

```python
# Create a dict object to keep track of the tweets
dict_ = {'user': [], 'date': [], 'text': [], 'favorite_count': [],'location':[],'hashtags':[]}

# Run the search API call
for status in python_tweets.search(**query)['statuses']:
    # Add the data we would like
    dict_['hashtags'].append(status['entities']['hashtags'])
    dict_['user'].append(status['user']['screen_name'])
    dict_['location'].append(status['user']['location'])
    dict_['date'].append(status['created_at'])
    dict_['text'].append(status['text'])
    dict_['favorite_count'].append(status['favorite_count'])
```

```python
import time

# Keep collecting the tweets
for r in range(0,360):
    for status in python_tweets.search(**query)['statuses']:
        dict_['hashtags'].append(status['entities']['hashtags'])
        dict_['user'].append(status['user']['screen_name'])
        dict_['location'].append(status['user']['location'])
        dict_['date'].append(status['created_at'])
        dict_['text'].append(status['text'])
        dict_['favorite_count'].append(status['favorite_count'])
    time.sleep(10) # Wait 10 seconds between requests
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent='User-Agent')
# Go through all tweets and add locations to 'coordinates' dictionary
coordinates = {'latitude': [], 'longitude': [],'country':[]}
for count, user_loc in enumerate(df.location):
    location = geolocator.geocode(user_loc)
    # If coordinates are found for location
    if location:
        coordinates['latitude'].append(location.latitude)
        coordinates['longitude'].append(location.longitude)
        coordinates['country'].append(location.raw['display_name']
            .rsplit(',',1)[-1]) 
```
